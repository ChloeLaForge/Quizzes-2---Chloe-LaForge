```python
female_passengers = titanic[titanic['sex'] == 'female']
num_female_survived = female_passengers['survived'].sum()
total_female_passengers = len(female_passengers)

percentage_female_survived = (num_female_survived / total_female_passengers) * 100
rounded_percentage = round(percentage_female_survived)

print(f"Percentage of female passengers who survived: {rounded_percentage}%")
```

```python
import pandas as pd

# Calculate the correlation matrix for numerical columns
correlation_matrix = iris.select_dtypes(include=['number']).corr()

# Exclude self-correlations and duplicate pairs by taking the upper triangle
corr_unstacked = correlation_matrix.unstack()
sorted_corr = corr_unstacked.sort_values(kind="quicksort", ascending=False)

# Filter out self-correlations (correlation of a variable with itself is always 1)
# and take only the upper triangle to avoid duplicate pairs (e.g., A-B and B-A)
positive_correlations = sorted_corr[sorted_corr < 1]

# Get the highest positive correlation pair
highest_positive_corr_pair = positive_correlations.drop_duplicates().iloc[[0]]

print("Variable pair with the highest Pearson correlation:")
print(highest_positive_corr_pair)
```
